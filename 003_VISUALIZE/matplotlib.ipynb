## Load Dataset

```python
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
```

```python
download_root = './data'
```

```python
batch_size = 20
```

```python
# Normalize data with mean=0.1307, std=0.3081
mnist_transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
```

```python
train_dataset = MNIST(download_root, transform=mnist_transform, train=True, download=True)
```

```python
train_loader = DataLoader(dataset=train_dataset, 
                         batch_size=batch_size)
```

## MNIST Dataset Visualize

```python
x_train, target = next(iter(train_loader))
x_train.shape, target.shape
x_train = x_train.view(batch_size, 28, 28)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_size_inches(10, 10)

for idx, x in enumerate(x_train):
    plt.subplot(4, 5, idx+1)
    x = x.reshape(28, 28)
    plt.imshow(x, cmap='gray', interpolation='none')
    plt.title("{}".format(target[idx]))
    plt.xticks([])
    plt.yticks([])
```
